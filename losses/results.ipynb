```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
import numpy as np
```

```python
# Trained on my PC
#t_50 = [1.25, 0.98, 0.90, 0.86, 0.83]
t_50 = [1.25, 0.98, 0.90]
#v_50 = [1.14, 1.04, 1.00, 0.97, 0.96] 
v_50 = [1.14, 1.04, 1.00] 
t_101 = [1.26, 0.98, 0.91]
v_101 = [1.14, 1.05, 1.00]  
```

```python
plt.scatter(np.arange(len(v_50)), v_50, label = "Validation", color = 'blue')
plt.scatter(np.arange(len(t_50)), t_50,  label = "Training", color = 'orange')
plt.scatter(np.arange(len(v_101)), v_101, label = "Validation", color = 'blue')
plt.scatter(np.arange(len(t_101)), t_101,  label = "Training", color = 'orange')
plt.plot(np.arange(len(v_50)), v_50, ls = ':', color = 'blue')
plt.plot(np.arange(len(t_50)), t_50, ls = ':', color = 'orange')
plt.plot(np.arange(len(v_101)), v_101, ls = '-.', color = 'blue')
plt.plot(np.arange(len(t_101)), t_101, ls = '-.', color = 'orange')
plt.title('CE during training')
plt.xlabel('Epoch')
plt.ylabel('Cross Entropy Loss')

legend_elements = [
    Line2D([0], [0], linestyle=':', label='ResNet 50'),
    Line2D([0], [0], linestyle='-.', label='ResNet 101'),
    Patch(facecolor='blue', label='Validation'),
    Patch(facecolor='orange', label='Training')
]
plt.legend(handles=legend_elements)
```

```python
def plot(val, training, title):
    plt.scatter(np.arange(len(val)), val, label = "Validation")
    plt.scatter(np.arange(len(training)), training,  label = "Training")
    plt.plot(np.arange(len(val)), val, ls = ':')
    plt.plot(np.arange(len(training)), training, ls = ':')
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Cross Entropy Loss')
    plt.legend()
```

```python
plot(v_50, t_50, "Metrics using ResNet 50")
```

```python
plot(v_101, t_101, "Metrics using ResNet 101")
```
